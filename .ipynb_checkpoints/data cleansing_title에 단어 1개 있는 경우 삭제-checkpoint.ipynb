## 중복 Title 처리 --> data_duplication

- case2) **current_status_of_industry:** '업계현황에 대한 레포트'이기 때문에 삭제 

- case1) **title_duplication:** 

>'title'과 'coname'열을 기준으로 중복 obs들 확인하기  --> Title 자체는 올바르게 들어갔으므로 중복된 것을 하나로 만들기 

```python
title_dup = df.duplicated(['title','coname'], keep = False) 
# df.duplicated(): 'title'과 'coname'기준으로 중복행을 찾음
#  keep = False 중복행들 모두 True --> df.duplicated():중복행을 남김 

df_title_dup = df.loc[title_dup==True, ]
df_title_dup
# df_title_dup.to_csv("./checklist data for data cleansing/(title_duplication) anal_reprt_title_car.csv", index=False, encoding = 'cp949')
```

```python
#  'title'과 'coname'기준으로 중복행인 obs들 중 첫번째만 남기고 제거 
df = df.drop_duplicates(['title','coname'], keep = 'first') 
# keep = 'first' 중복행들 중 첫번째 행 F, 중복행이 아니어도 F --> 중복행들 중 첫번째 이외의 행만 TRUE 
# df.drop_duplicated(): TRUE인 중복행(첫번째 이외의 행)을 drop함 --> False인 행만 살아남음

df
```

## Title 내에 등장하는 기업명 제거 
- 'title'에서 "-"를 찾고, "-"가 존재하는 문자열의 위치=ind 를 찾는다. 
- 'title' 문자열을 ind 까지 인덱싱해서 제거할 문자열=rm을 정의하고 ""로 replace 한다.  

```python
# 기업명 제거 함수 정의 
def cleansing_coname(title):
    if title.find("-") != -1: # 기업명이 존재한다면 # 해당단어가 존재하지 않으면, text.find(i) = -1 
        ind = title.index("-")
        rm = title[:ind+1] # 제거할 문자열 
        title = title.replace(rm,"").strip()
    return title

# 기업명 제거 
df['title']=df['title'].apply(lambda x: cleansing_coname(x))

df
# df.to_csv("./checklist data for data cleansing/(기업명제거) anal_reprt_title_car.csv",index=False, encoding = 'cp949')
```

## sentiment_analysis_by_huggingface_model

```python
!pip install transformers
!pip install tensorflow
!pip install torch
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
import tensorflow as tf
import torch
import numpy as np
```

```python
def new_softmax(a) : 
    c = np.max(a) # 최댓값
    exp_a = np.exp(a-c) # 각각의 원소에 최댓값을 뺀 값에 exp를 취한다. (이를 통해 overflow 방지)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
model_name = "snunlp/KR-FinBert-SC"

tokenizer = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")
model = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
```

```python
classifier = pipeline("feature-extraction", model=model, tokenizer=tokenizer)
```

```python
df['label'] = df['title'].apply(lambda x: new_softmax(classifier(x)))
```

```python
df['neg_prob']=df['label'].apply(lambda x: x[0,0]) # x[0,0] array indexing 
df['neu_prob']=df['label'].apply(lambda x: x[0,1])
df['pos_prob']=df['label'].apply(lambda x: x[0,2])
df
```

```python
classifier2 = pipeline("text-classification", model=model, tokenizer=tokenizer)  # 부정,중립,긍정 확률을 계산한 후 가장 확률이 큰 쪽으로 분류하고 그 확률을 스코어로 출력
```

```python
df['label']=df['title'].apply(lambda x: classifier2(x))
```

```python
df
```

```python
label_list = df['label'].apply(lambda x: x[0].get('label')) # list의 정수 인덱스 0번을 가져오고, 딕셔너리의 'label' 값 가져오기 

print('전체 label 개수: ',len(label_list))

neg_num = (label_list == 'negative').sum()
print('\nnegative의 개수: ',neg_num)
print('negative가 차지하는 비율: ',neg_num/len(label_list))

neu_num = (label_list == 'neutral').sum()
print('\nneutral의 개수: ',neu_num)
print('neutral이 차지하는 비율: ',neu_num/len(label_list))

pos_num = (label_list == 'positive').sum()
print('\npositive의 개수: ',pos_num)
print('positive가 차지하는 비율: ',pos_num/len(label_list))
```
