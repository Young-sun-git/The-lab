```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
import tensorflow as tf
import torch
import numpy as np
```

```python
def new_softmax(a) : 
    c = np.max(a) # 최댓값
    exp_a = np.exp(a-c) # 각각의 원소에 최댓값을 뺀 값에 exp를 취한다. (이를 통해 overflow 방지)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
model_name = "snunlp/KR-FinBert-SC"
```

```python
tokenizer = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")
model = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
```

```python
classifier = pipeline("feature-extraction", model=model, tokenizer=tokenizer)
```

```python
new_softmax(classifier("부품 공급 차질에…기아차 광주공장 전면 가동 중단")) # 부정적 뉴스 # (부정확률,중립확률,긍정확률) (합하면 1)
```

```python
new_softmax(classifier("박문덕 하이트진로 회장 호시우행으로 100년 기업 준비")) # 중립적 뉴스
```

```python
new_softmax(classifier("하이트진로, 1분기 영업익 262억…전년비 148% 상승")) # 긍정적 뉴스
```

```python
classifier2 = pipeline("text-classification", model=model, tokenizer=tokenizer)  # 부정,중립,긍정 확률을 계산한 후 가장 확률이 큰 쪽으로 분류하고 그 확률을 스코어로 출력
```

```python
classifier2("대림산업, 인천 'e편한세상 서창' 분양")
```
