```python
df=df_arCar.loc[:,['code','analyst','coname','carm','yy','mm','dd','title']]
df.tail(100)
```

## 중복 title 추출 

```python
title_list = list(set(df['title']))  # set은 허용 X
# print(title_list)
new_df = pd.DataFrame()

for i in title_list:
    tmp = df[df['title'] == i]
    if len(tmp) >= 2:
        new_df = pd.concat([new_df, tmp])
```

```python
new_df
```

```python
new_df.to_csv('./new_df.csv', encoding='cp949')
```
