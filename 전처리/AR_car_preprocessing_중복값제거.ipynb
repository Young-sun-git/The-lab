## title 안의 기업명 제거 

```python
# 기업명 
company=list(set(removeDuplication_arCarm['coname']))# uniqe한 기업명 
```

```python
# 기업명 제거 함수 정의 
def cleansing(title):
    if title.find("-") != -1: # 기업명이 존재한다면 # 해당단어가 존재하지 않으면, text.find(i) = -1 
        ind = title.index("-")
        rm = title[:ind+1]
        title = title.replace(rm,"").strip()
    return title

# 기업명 제거 
removeDuplication_arCarm['title']=removeDuplication_arCarm['title'].apply(lambda x: cleansing(x))

removeDuplication_arCarm['title'].head(20)
```

```python
ARCarm=removeDuplication_arCarm
```

## title 전처리 


```python
# title 전처리 
# 한글과 공백을 제하고 모두 제거 
ARCarm['title']=ARCarm['title'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

# 데이터 정제후 빈문자열(na 로 바꾼 후) 제거 
ARCarm['title'].replace('',np.nan,inplace=True)
# print(text.isnull().sum())

# na 제거
ARCarm = ARCarm.dropna(subset=['title'])
# print(len(ARCarm))
# print(ARCarm.isnull().sum())
```
